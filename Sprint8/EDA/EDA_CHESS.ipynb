```python
'''
Formulación de Hipótesis
Formularemos varias hipótesis para analizar:

Hipótesis 1: Los jugadores con una calificación más alta tienden a ganar más partidas.
Vamos a comparar las calificaciones promedio de los jugadores ganadores (blanco y negro) y perdedores.
Hipótesis 2: Las partidas con un mayor número de turnos tienden a terminar en tablas.
Vamos a analizar la distribución de la cantidad de turnos para diferentes estados de victoria.
Hipótesis 3: Ciertas aperturas específicas tienen una mayor tasa de victorias para las blancas o las negras.
Vamos a calcular las tasas de victoria para las aperturas más comunes.
Hipótesis 4: Los juegos con incrementos de tiempo más largos tienden a tener más turnos.
Vamos a correlacionar el incremento de tiempo con la cantidad de turnos en las partidas.
Hipótesis 5: Los jugadores que usan incrementos de tiempo más largos tienen una mayor calificación promedio.
Hipótesis 6: El resultado de la partida (victoria de blancas o negras) depende del tipo de apertura utilizado.
'''
```

```python
'''
Hipótesis 1: Los jugadores con una calificación más alta tienden a ganar más partidas.
'''
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt

# Filtrar datos de ganadores y perdedores
winners = chess_data_cleaned[chess_data_cleaned['winner'] != 'draw']
losers = chess_data_cleaned[chess_data_cleaned['winner'] != 'draw']

# Calificaciones de ganadores y perdedores
white_winners = winners[winners['winner'] == 'white']['white_rating']
black_winners = winners[winners['winner'] == 'black']['black_rating']
white_losers = losers[losers['winner'] == 'black']['white_rating']
black_losers = losers[losers['winner'] == 'white']['black_rating']

# Prueba t para ganadores vs perdedores
t_stat_white, p_val_white = ttest_ind(white_winners, white_losers, equal_var=False)
t_stat_black, p_val_black = ttest_ind(black_winners, black_losers, equal_var=False)

u_stat_white, p_valor_white = mannwhitneyu(white_winners, white_losers)
u_stat_black, p_valor_black = mannwhitneyu(black_winners, black_losers)

# Visualización con Seaborn
fig, axs = plt.subplots(1,2,figsize=(20, 10))

# Calificaciones de los ganadores
average_ratings_winners_white = winners.groupby('winner')[['white_rating']].mean().reset_index()
average_ratings_winners_black = winners.groupby('winner')[['black_rating']].mean().reset_index()
sns.barplot(x='winner', y='white_rating',  data=average_ratings_winners_white,hue= "winner" ,alpha=0.7, legend=True, palette='dark:steelblue', ax=axs[0])
sns.barplot(x='winner', y='black_rating', data=average_ratings_winners_black,hue= "winner" ,alpha=0.7, legend=True, palette='dark:steelblue',ax=axs[1])



axs[0].set_title('Calificaciones Promedio cuando Ganan Blancas')
axs[0].set_ylabel('Calificación Promedio',loc="center")
axs[0].set_xlabel('Tipo Piezas',loc="center")
axs[1].set_title('Calificaciones Promedio Cuando Ganan negras')
axs[1].set_ylabel('Calificación Promedio',loc="center")
axs[1].set_xlabel('Tipo Piezas',loc="center")
axs[0].legend()
axs[1].legend()
plt.tight_layout()

print(f"Prueba t para jugadores blancos: t-stat={t_stat_white}, p-value={p_val_white}")
print(f"Prueba t para jugadores negros: t-stat={t_stat_black}, p-value={p_val_black}")
print(f"Prueba t para jugadores blancos: u-stat={u_stat_white}, p-value={p_valor_white}")
print(f"Prueba t para jugadores negros: u-stat={u_stat_black}, p-value={p_valor_black}")
```

```python
'''
Hipótesis 3: Las partidas con un mayor número de turnos tienden a terminar en tablas.
'''

from scipy.stats import f_oneway

# Filtrar turnos por estado de victoria
turns_draw = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'draw']['turns']
turns_mate = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'mate']['turns']
turns_outoftime = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'outoftime']['turns']
turns_resign = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'resign']['turns']

# Prueba ANOVA
f_stat, p_val = f_oneway(turns_draw, turns_mate, turns_outoftime, turns_resign)

# Visualización con Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='victory_status', y='turns', data=chess_data_cleaned, showmeans=True)
plt.title('Numero de Turnos por tipo de Resultado')
plt.xlabel('Tipo de Resultado')
plt.ylabel('Cantidad de Turnos')
plt.show()
plt.tight_layout()

print(f"Prueba ANOVA: f-stat={f_stat}, p-value={p_val}")
```

```python
'''
Hipótesis 2: Ciertas aperturas específicas tienen una mayor tasa de victorias para las blancas o las negras.


'''
from scipy.stats import chi2_contingency

# Focalizándonos en las 10 aperturas más comunes
common_openings = chess_data_cleaned['opening_name'].value_counts().head(15).index
top_openings_data = chess_data_cleaned[chess_data_cleaned['opening_name'].isin(common_openings)]

# Calculando las tasas de victoria para blancas y negras
opening_win_rates = top_openings_data.groupby(['opening_name', 'winner']).size().unstack().fillna(0)

# Prueba de chi-cuadrado
chi2, p_val, dof, ex = chi2_contingency(opening_win_rates)

# Visualización con Seaborn
fig, axs = plt.subplots(2,1,figsize=(40, 20))
sns.barplot(data=opening_win_rates[['white', 'black']].reset_index(), x='opening_name', y='white', color='steelblue', label='White Wins',alpha = 0.5, ax=axs[0])
sns.barplot(data=opening_win_rates[['white', 'black']].reset_index(), x='opening_name', y='black', color='steelblue', label='Black Wins',alpha = 0.5,ax=axs[1])
axs[0].set_title('Tasas de Victoria para las 10 Aperturas Más Comunes')
axs[0].set_xlabel('Apertura' )
axs[0].set_ylabel('Número de Victorias')
axs[0].tick_params(axis='x', rotation=45)
axs[1].set_title('Tasas de Victoria para las 10 Aperturas Más Comunes')
axs[1].set_xlabel('Apertura')
axs[1].set_ylabel('Número de Victorias')
axs[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

print(f"Prueba de chi-cuadrado: chi2={chi2}, p-value={p_val}, grados de libertad {dof}")
```

```python
'''
Hipótesis 4: Los juegos con incrementos de tiempo más largos tienden a tener más turnos.
Vamos a correlacionar el incremento de tiempo con la cantidad de turnos en las partidas.
'''

from scipy.stats import pearsonr

# Convertir el incremento de tiempo a segundos
def increment_to_seconds(increment):
    minutes, seconds = map(int, increment.split('+'))
    return minutes * 60 + seconds

chess_data_cleaned['increment_seconds'] = chess_data_cleaned['increment_code'].apply(increment_to_seconds)

# Prueba de correlación de Pearson
corr, p_val = pearsonr(chess_data_cleaned['increment_seconds'], chess_data_cleaned['turns'])

# Visualización
plt.figure(figsize=(14, 6))
plt.scatter(chess_data_cleaned['increment_seconds'], chess_data_cleaned['turns'], alpha=0.5)
plt.title('Relación entre Incremento de Tiempo y Cantidad de Turnos')
plt.xlabel('Incremento de Tiempo (segundos)')
plt.ylabel('Cantidad de Turnos')
plt.show()

print(f"Prueba de correlación de Pearson: correlación={corr}, p-value={p_val}")
```

```python
'''
Hipótesis 5: Los jugadores que usan incrementos de tiempo más largos tienen una mayor calificación promedio.
'''

from scipy.stats import ttest_ind

# Convertir el incremento de tiempo a segundos
def increment_to_seconds(increment):
    minutes, seconds = map(int, increment.split('+'))
    return minutes * 60 + seconds

chess_data_cleaned['increment_seconds'] = chess_data_cleaned['increment_code'].apply(increment_to_seconds)

# Dividir en grupos de incrementos largos y cortos (usaremos la mediana como punto de corte)
median_increment = chess_data_cleaned['increment_seconds'].median()
long_increment = chess_data_cleaned[chess_data_cleaned['increment_seconds'] > median_increment]
short_increment = chess_data_cleaned[chess_data_cleaned['increment_seconds'] <= median_increment]

# Calificaciones promedio para ambos grupos
avg_rating_long = long_increment[['white_rating', 'black_rating']].mean().mean()
avg_rating_short = short_increment[['white_rating', 'black_rating']].mean().mean()

# Prueba t
t_stat, p_val = ttest_ind(long_increment[['white_rating', 'black_rating']].values.flatten(),
                          short_increment[['white_rating', 'black_rating']].values.flatten(), equal_var=False)

# Visualización con Seaborn
plt.figure(figsize=(200, 6))
sns.boxplot(x='increment_seconds', y='white_rating', data=chess_data_cleaned, showmeans=True)
plt.axhline(avg_rating_long, color='r', linestyle='--', label='Avg Long Increment')
plt.axhline(avg_rating_short, color='b', linestyle='--', label='Avg Short Increment')
plt.title('Calificación Promedio de Jugadores por Incremento de Tiempo')
plt.xlabel('Incremento de Tiempo (segundos)')
plt.ylabel('Calificación Promedio')
plt.legend()
plt.show()

print(f"Calificación Promedio con Incremento Largo: {avg_rating_long}")
print(f"Calificación Promedio con Incremento Corto: {avg_rating_short}")
print(f"Prueba t: t-stat={t_stat}, p-value={p_val}")
```

```python
'''
Hipótesis 6: El resultado de la partida (victoria de blancas o negras) depende del tipo de apertura utilizado.

'''

from scipy.stats import chi2_contingency

# Focalizándonos en las 10 aperturas más comunes
common_openings = chess_data_cleaned['opening_name'].value_counts().head(15).index
top_openings_data = chess_data_cleaned[chess_data_cleaned['opening_name'].isin(common_openings)]

# Resultado de la partida según la apertura
opening_results = top_openings_data.groupby(['opening_name', 'winner']).size().unstack().fillna(0)

# Prueba de chi-cuadrado
chi2, p_val, dof, ex = chi2_contingency(opening_results)

# Visualización con Seaborn
plt.figure(figsize=(14, 8))
sns.heatmap(opening_results, annot=True, fmt="d", cmap="viridis")
plt.title('Resultados de Partida por Apertura')
plt.xlabel('Resultado')
plt.ylabel('Apertura')
plt.show()

print(f"Prueba de chi-cuadrado: chi2={chi2}, p-value={p_val}, Grados de libertad {dof}")
```

```python
'''
Hipótesis 7: Las partidas clasificadas (rated) tienden a durar más turnos que las no clasificadas.
'''

from scipy.stats import ttest_ind

# Dividir en partidas clasificadas y no clasificadas
rated_games = chess_data_cleaned[chess_data_cleaned['rated'] == True]
unrated_games = chess_data_cleaned[chess_data_cleaned['rated'] == False]

# Prueba t
t_stat, p_val = ttest_ind(rated_games['turns'], unrated_games['turns'], equal_var=False)

# Visualización con Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='rated', y='turns', data=chess_data_cleaned, showmeans=True)
plt.title('Duración de Partidas Clasificadas vs No Clasificadas')
plt.xlabel('Partida Clasificada')
plt.ylabel('Número de Turnos')
plt.show()

print(f"Prueba t: t-stat={t_stat}, p-value={p_val}")
```

```python
'''
Hipótesis 8: Las partidas con una mayor cantidad de movimientos tienden a ser ganadas por blancas.
'''

# Contar el número de movimientos en cada partida
chess_data_cleaned['move_count'] = chess_data_cleaned['moves'].apply(lambda x: len(x.split()))

# Filtrar datos de partidas ganadas por blancas y negras
white_wins = chess_data_cleaned[chess_data_cleaned['winner'] == 'white']['move_count']
black_wins = chess_data_cleaned[chess_data_cleaned['winner'] == 'black']['move_count']

# Prueba t para comparar la cantidad de movimientos
t_stat_8, p_val_8 = ttest_ind(white_wins, black_wins, equal_var=False)

# Visualización con Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='winner', y='move_count', data=chess_data_cleaned[chess_data_cleaned['winner'].isin(['white', 'black'])], showmeans=True)
plt.title('Cantidad de Movimientos en Partidas Ganadas por Blancas vs Negras')
plt.xlabel('Ganador')
plt.ylabel('Cantidad de Movimientos')
plt.show()

print(f"Prueba t: t-stat={t_stat_8}, p-value={p_val_8}")
```

```python
'''
Hipótesis 9: Las partidas que terminan en tablas tienen una mayor variedad de movimientos.
'''

def move_diversity(moves):
    move_list = moves.split()
    unique_moves = set(move_list)
    return len(unique_moves) / len(move_list)

chess_data_cleaned['move_diversity'] = chess_data_cleaned['moves'].apply(move_diversity)

# Filtrar datos de partidas que terminan en tablas y otras
draws = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'draw']['move_diversity']
non_draws = chess_data_cleaned[chess_data_cleaned['victory_status'] != 'draw']['move_diversity']

# Prueba t para comparar la diversidad de movimientos
t_stat_9, p_val_9 = ttest_ind(draws, non_draws, equal_var=False)

# Visualización con Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='victory_status', y='move_diversity', data=chess_data_cleaned, showmeans=True)
plt.title('Diversidad de Movimientos en Partidas que Terminan en Tablas vs Otras')
plt.xlabel('Estado de Victoria')
plt.ylabel('Diversidad de Movimientos')
plt.show()

print(f"Prueba t: t-stat={t_stat_9}, p-value={p_val_9}")
```

```python
'''
Hipótesis 10: Las partidas ganadas por jaque mate tienden a tener una mayor cantidad de movimientos de peones.
'''
def pawn_moves(moves):
    move_list = moves.split()
    pawn_moves_count = sum(1 for move in move_list if move[0] in 'abcdefgh')
    return pawn_moves_count

chess_data_cleaned['pawn_move_count'] = chess_data_cleaned['moves'].apply(pawn_moves)

# Filtrar datos de partidas ganadas por jaque mate y otras
mate_wins = chess_data_cleaned[chess_data_cleaned['victory_status'] == 'mate']['pawn_move_count']
non_mate_wins = chess_data_cleaned[chess_data_cleaned['victory_status'] != 'mate']['pawn_move_count']

print(sns.histplot(x=mate_wins, kde=True, color='violet',alpha =0.5))
print(sns.histplot(x=non_mate_wins, kde=True,color = 'black',alpha =0.5))

# Prueba t para comparar la cantidad de movimientos de peones, ya que ambas tienen una distribucion aprecida a la normal
t_stat_10, p_val_10 = ttest_ind(mate_wins, non_mate_wins, equal_var=False)
u_stat_10, up_val_10 = mannwhitneyu(mate_wins, non_mate_wins)

# Visualización con Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='victory_status', y='pawn_move_count', data=chess_data_cleaned, showmeans=True)
plt.title('Cantidad de Movimientos de Peones en Partidas Ganadas por Jaque Mate vs Otras')
plt.xlabel('Estado de Victoria')
plt.ylabel('Cantidad de Movimientos de Peones')
plt.show()

print(f"Prueba t: t-stat={t_stat_10}, p-value={p_val_10}")
print(f"Prueba t: MU-stat={u_stat_10}, p-value={up_val_10}")
```
